```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from jiwer import wer
from IPython.display import Audio, display
import os
import random


DATA_DIR = "/kaggle/input/ljspeech/LJSpeech-1.1"

AUDIO_DIR = DATA_DIR + "/wavs/"
META_FILE = DATA_DIR + "/metadata.csv"

alphabet = [ch for ch in "abcdefghijklmnopqrstuvwxyz'?! "]

FRAME_LEN = 256
FRAME_STEP = 160
FFT_SIZE = 256
BATCH = 32
EPOCHS = 20


metadata = pd.read_csv(
    META_FILE, 
    sep="|", 
    header=None, 
    quoting=3,
    names=["clip_id", "text_raw", "text_normalized"]
)

metadata = metadata[["clip_id", "text_normalized"]]
metadata = metadata.sample(frac=1).reset_index(drop=True)
metadata.head()

split_idx = int(len(metadata) * 0.90)

train_df = metadata[:split_idx]
valid_df = metadata[split_idx:]


print("Train samples:", len(train_df))
print("Validation samples:", len(valid_df))

char_to_num = keras.layers.StringLookup(vocabulary=alphabet, oov_token="")
num_to_char = keras.layers.StringLookup(
    vocabulary=char_to_num.get_vocabulary(),
    invert=True,
    oov_token=""
)

print("Vocabulary:", char_to_num.get_vocabulary())
print("Vocabulary size:", char_to_num.vocabulary_size())


def preprocess_sample(file_id, transcript):
    wav_bytes = tf.io.read_file(AUDIO_DIR + file_id + ".wav")
    audio, _ = tf.audio.decode_wav(wav_bytes)
    audio = tf.squeeze(audio, axis=-1)
    audio = tf.cast(audio, tf.float32)

    spectro = tf.signal.stft(
        audio,
        frame_length=FRAME_LEN,
        frame_step=FRAME_STEP,
        fft_length=FFT_SIZE
    )
    spectro = tf.abs(spectro) ** 0.5

    mean = tf.math.reduce_mean(spectro, axis=1, keepdims=True)
    std = tf.math.reduce_std(spectro, axis=1, keepdims=True)
    spectro = (spectro - mean) / (std + 1e-10)

    transcript = tf.strings.lower(transcript)
    chars = tf.strings.unicode_split(transcript, "UTF-8")
    encoded = char_to_num(chars)

    return spectro, encoded



train_ds = tf.data.Dataset.from_tensor_slices(
    (train_df["clip_id"], train_df["text_normalized"])
)
train_ds = (
    train_ds.map(preprocess_sample, num_parallel_calls=tf.data.AUTOTUNE)
            .padded_batch(BATCH)
            .prefetch(tf.data.AUTOTUNE)
)

valid_ds = tf.data.Dataset.from_tensor_slices(
    (valid_df["clip_id"], valid_df["text_normalized"])
)
valid_ds = (
    valid_ds.map(preprocess_sample, num_parallel_calls=tf.data.AUTOTUNE)
            .padded_batch(BATCH)
            .prefetch(tf.data.AUTOTUNE)
)

train_ds, valid_ds


for s_batch, t_batch in train_ds.take(1):
    spec = s_batch[0].numpy()
    text = t_batch[0]

    spec_vis = np.transpose(spec)
    spec_vis = np.array([np.trim_zeros(row) for row in spec_vis])

    decoded = tf.strings.reduce_join(num_to_char(text)).numpy().decode("utf-8")

    plt.figure(figsize=(9,5))
    plt.imshow(spec_vis, aspect="auto", origin="lower")
    plt.title(decoded)
    plt.colorbar()
    plt.show()

    wav_file = AUDIO_DIR + train_df.iloc[0]["clip_id"] + ".wav"
    wav_bytes = tf.io.read_file(wav_file)
    wav_data, _ = tf.audio.decode_wav(wav_bytes)
    
    audio_np = wav_data.numpy().flatten()
    audio_np = audio_np / (np.max(np.abs(audio_np)) + 1e-9)
    
    display(Audio(audio_np, rate=22050))



def CTC_Loss(y_true, y_pred):
    batch_len = tf.cast(tf.shape(y_true)[0], tf.int64)
    time_steps = tf.cast(tf.shape(y_pred)[1], tf.int64)
    label_len = tf.cast(tf.shape(y_true)[1], tf.int64)

    input_length = time_steps * tf.ones((batch_len, 1), dtype=tf.int64)
    label_length = label_len * tf.ones((batch_len, 1), dtype=tf.int64)

    return keras.backend.ctc_batch_cost(y_true, y_pred, input_length, label_length)



def make_model(feature_dim, vocab_size, rnn_count=3, rnn_units=256):
    inp = layers.Input(shape=(None, feature_dim), name="spectrogram")

    x = layers.Reshape((-1, feature_dim, 1))(inp)

    x = layers.Conv2D(32, (11, 41), strides=(2, 2), padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv2D(32, (11, 21), strides=(1, 2), padding="same", use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Reshape((-1, x.shape[-2] * x.shape[-1]))(x)

    for idx in range(rnn_count):
        gru = layers.GRU(
            rnn_units,
            activation="tanh",
            recurrent_activation="sigmoid",
            return_sequences=True,
            reset_after=True,
        )
        x = layers.Bidirectional(gru)(x)
        if idx < rnn_count - 1:
            x = layers.Dropout(0.5)(x)

    x = layers.Dense(rnn_units * 2)(x)
    x = layers.ReLU()(x)
    x = layers.Dropout(0.5)(x)

    out = layers.Dense(vocab_size + 1, activation="softmax")(x)

    model = keras.Model(inp, out)
    model.compile(optimizer=keras.optimizers.Adam(1e-4), loss=CTC_Loss)

    return model


speech_model = make_model(
    feature_dim=FFT_SIZE // 2 + 1,
    vocab_size=char_to_num.vocabulary_size(),
    rnn_units=256
)

speech_model.summary()



def decode_predictions(pred_batch):
    time_len = np.ones(pred_batch.shape[0]) * pred_batch.shape[1]
    decoded, _ = keras.backend.ctc_decode(pred_batch, input_length=time_len, greedy=True)
    decoded = decoded[0]

    texts = []
    for seq in decoded:
        text = tf.strings.reduce_join(num_to_char(seq)).numpy().decode("utf-8")
        texts.append(text)
    return texts



class ValidationCallback(keras.callbacks.Callback):
    def __init__(self, ds):
        super().__init__()
        self.ds = ds

    def on_epoch_end(self, epoch, logs=None):
        preds, trues = [], []

        for xb, yb in self.ds:
            out = self.model.predict(xb)
            decoded = decode_predictions(out)
            preds.extend(decoded)

            for lab in yb:
                txt = tf.strings.reduce_join(num_to_char(lab)).numpy().decode("utf-8")
                trues.append(txt)

        score = wer(trues, preds)
        print("\n" + "="*80)
        print(f"WER after epoch {epoch+1}: {score:.4f}")
        print("="*80)


val_cb = ValidationCallback(valid_ds)

history = speech_model.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=EPOCHS,
    callbacks=[val_cb]
)


final_preds = []
final_trues = []

for xb, yb in valid_ds:
    out = speech_model.predict(xb)
    decoded = decode_predictions(out)
    final_preds.extend(decoded)

    for lab in yb:
        txt = tf.strings.reduce_join(num_to_char(lab)).numpy().decode("utf-8")
        final_trues.append(txt)

final_wer = wer(final_trues, final_preds)
print("\nFINAL WER:", final_wer)


def recognize_wav_file(path):
    audio_bin = tf.io.read_file(path)
    audio, _ = tf.audio.decode_wav(audio_bin)
    audio = tf.squeeze(audio, axis=-1)
    audio = tf.cast(audio, tf.float32)

    spectrogram = tf.signal.stft(
        audio,
        frame_length=FRAME_LEN,
        frame_step=FRAME_STEP,
        fft_length=FFT_SIZE,
    )
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.math.pow(spectrogram, 0.5)

    means = tf.math.reduce_mean(spectrogram, axis=1, keepdims=True)
    stddevs = tf.math.reduce_std(spectrogram, axis=1, keepdims=True)
    spectrogram = (spectrogram - means) / (stddevs + 1e-10)

    spectrogram = tf.expand_dims(spectrogram, axis=0)

    pred = speech_model.predict(spectrogram)
    text = decode_predictions(pred)[0]
    return text


def test_multiple_examples(model, df, n=5):
    print(f"\n=== ТЕСТ НА {n} ПРИКЛАДАХ ===\n")
    for i in range(n):
        row = df.iloc[random.randint(0, len(df) - 1)]

        wav_path = AUDIO_DIR + row["clip_id"] + ".wav"
        gt_text = row["text_normalized"]

        pred_text = recognize_wav_file(wav_path)

        example_wer = wer([gt_text], [pred_text])

        print(f"Приклад #{i+1}")
        print("GT   :", gt_text)
        print("PRED :", pred_text)
        print(f"WER  : {example_wer:.3f}")
        print("-" * 80)


test_multiple_examples(speech_model, valid_df, n=5)

```

```python
from spellchecker import SpellChecker
spell = SpellChecker()

def postprocess_spellchecker(text):
    words = text.split()
    corrected = []
    for w in words:
        if w.isalpha():
            corr = spell.correction(w)
            if corr is None:
                corr = w
            corrected.append(corr)
        else:
            corrected.append(w)
    return " ".join(corrected)


from autocorrect import Speller
auto_spell = Speller(lang='en')

def postprocess_autocorrect(text):
    return auto_spell(text)

from happytransformer import HappyTextToText, TTSettings

grammar_model = HappyTextToText("T5", "vennify/t5-base-grammar-correction")

grammar_args = TTSettings(
    num_beams=5,
    min_length=1,
    early_stopping=True,
    no_repeat_ngram_size=3
)

def postprocess_t5(text):
    result = grammar_model.generate_text(text, args=grammar_args)
    return result.text

def evaluate_postprocessing(model, df, n=20):
    print(f"\n=== Оцінка пост-обробки на {n} прикладах ===\n")

    wer_raw = []
    wer_spell = []
    wer_auto = []
    wer_t5 = []

    for i in range(n):
        row = df.iloc[random.randint(0, len(df)-1)]
        wav_path = AUDIO_DIR + row["clip_id"] + ".wav"
        gt = row["text_normalized"]

        raw_pred = recognize_wav_file(wav_path)

        spell_pred = postprocess_spellchecker(raw_pred)
        auto_pred = postprocess_autocorrect(raw_pred)
        t5_pred   = postprocess_t5(raw_pred)

        wer_raw.append(wer([gt],[raw_pred]))
        wer_spell.append(wer([gt],[spell_pred]))
        wer_auto.append(wer([gt],[auto_pred]))
        wer_t5.append(wer([gt],[t5_pred]))

    print("RAW WER       :", np.mean(wer_raw))
    print("SpellChecker  :", np.mean(wer_spell))
    print("Autocorrect   :", np.mean(wer_auto))
    print("T5 Correction :", np.mean(wer_t5))

evaluate_postprocessing(speech_model, valid_df, n=20)
```
