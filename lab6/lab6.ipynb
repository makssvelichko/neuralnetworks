```python
import os, random, warnings
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix, ConfusionMatrixDisplay,
                             precision_recall_fscore_support, accuracy_score)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

warnings.filterwarnings("ignore")

SEED = 42
random.seed(SEED); 
np.random.seed(SEED); 
tf.random.set_seed(SEED)

POS_DIR = "/kaggle/input/logo-dataset-vw/dataset_vw"
NEG_DIR = "/kaggle/input/dataset-back/dataset_background" 

IMG_SIZE = 299
BATCH_SIZE = 32
VAL_SPLIT = 0.2
TEST_SPLIT = 0.1
EPOCHS = 20

def list_images(folder):
    if not os.path.exists(folder): return []
    exts = (".jpg",".jpeg",".png",".webp")
    return [os.path.join(folder,f) for f in os.listdir(folder) if f.lower().endswith(exts)]

pos_files = list_images(POS_DIR)
neg_files = list_images(NEG_DIR)
print(f"Found {len(pos_files)} positive, {len(neg_files)} negative images")
assert len(pos_files) + len(neg_files) >= 300, "Потрібно мінімум 300 зображень."

X = pos_files + neg_files
y = [1]*len(pos_files) + [0]*len(neg_files)


X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y, test_size=TEST_SPLIT, random_state=SEED, stratify=y
)
val_rel = VAL_SPLIT / (1.0 - TEST_SPLIT)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=val_rel, random_state=SEED, stratify=y_trainval
)

import collections
def label_counts(labels, name):
    c = collections.Counter(labels)
    print(f"{name}: total={len(labels)} | zeros={c[0]} | ones={c[1]} | frac_pos={c[1]/max(1,len(labels)):.3f}")
label_counts(y_train, "TRAIN"); label_counts(y_val, "VAL"); label_counts(y_test, "TEST")

AUTO = tf.data.AUTOTUNE

def decode_and_resize(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE), method="lanczos3")
    img = (img - 0.5) * 2.0
    return img, tf.cast(label, tf.float32)

data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.07),
    layers.RandomZoom(0.1),
    layers.RandomTranslation(0.05, 0.05),
], name="aug")

def make_ds(paths, labels, shuffle=False, augment=False):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=min(len(paths), 2048), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(decode_and_resize, num_parallel_calls=AUTO)
    if augment:
        ds = ds.map(lambda x,y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTO)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTO)
    return ds

train_ds = make_ds(X_train, y_train, shuffle=True,  augment=True)
val_ds   = make_ds(X_val,   y_val,   shuffle=False, augment=False)
test_ds  = make_ds(X_test,  y_test,  shuffle=False, augment=False)


BN_MOMENTUM = 0.9
BN_EPS      = 1e-3

def SepConv_BN(x, filters, prefix, kernel_size=3, strides=1, activation=True):
    if strides == 2:
        x = layers.ZeroPadding2D(((1,1),(1,1)), name=prefix+"_pad")(x)
        padding = "valid"
    else:
        padding = "same"
    x = layers.SeparableConv2D(filters, kernel_size, strides=strides, padding=padding,
                               use_bias=False, name=prefix)(x)
    x = layers.BatchNormalization(axis=3, momentum=BN_MOMENTUM, epsilon=BN_EPS,
                                  scale=False, name=prefix+"_bn")(x)
    if activation:
        x = layers.Activation("relu", name=prefix+"_relu")(x)
    return x

def conv_bn(x, filters, kernel_size, strides=1, name_prefix=""):
    x = layers.Conv2D(filters, kernel_size, strides=strides, use_bias=False, padding="same",
                      name=name_prefix+"conv")(x)
    x = layers.BatchNormalization(axis=3, momentum=BN_MOMENTUM, epsilon=BN_EPS,
                                  scale=False, name=name_prefix+"bn")(x)
    x = layers.Activation("relu", name=name_prefix+"relu")(x)
    return x

def xception_build(input_shape=(IMG_SIZE, IMG_SIZE, 3), dropout=0.4):
    img_input = layers.Input(shape=input_shape)

    x = conv_bn(img_input, 32, 3, strides=2, name_prefix="block1_conv1_")
    x = conv_bn(x,         64, 3, strides=1, name_prefix="block1_conv2_")

    residual = layers.Conv2D(128, 1, strides=2, padding="same", use_bias=False, name="block2_res_conv")(x)
    residual = layers.BatchNormalization(axis=3, momentum=BN_MOMENTUM, epsilon=BN_EPS, scale=False, name="block2_res_bn")(residual)
    x = SepConv_BN(x, 128, "block2_sepconv1")
    x = SepConv_BN(x, 128, "block2_sepconv2", activation=False)
    x = layers.MaxPooling2D(3, strides=2, padding="same", name="block2_pool")(x)
    x = layers.add([x, residual], name="block2_add")

    residual = layers.Conv2D(256, 1, strides=2, padding="same", use_bias=False, name="block3_res_conv")(x)
    residual = layers.BatchNormalization(axis=3, momentum=BN_MOMENTUM, epsilon=BN_EPS, scale=False, name="block3_res_bn")(residual)
    x = layers.Activation("relu", name="block3_act")(x)
    x = SepConv_BN(x, 256, "block3_sepconv1")
    x = SepConv_BN(x, 256, "block3_sepconv2", activation=False)
    x = layers.MaxPooling2D(3, strides=2, padding="same", name="block3_pool")(x)
    x = layers.add([x, residual], name="block3_add")

    residual = layers.Conv2D(728, 1, strides=2, padding="same", use_bias=False, name="block4_res_conv")(x)
    residual = layers.BatchNormalization(axis=3, momentum=BN_MOMENTUM, epsilon=BN_EPS, scale=False, name="block4_res_bn")(residual)
    x = layers.Activation("relu", name="block4_act")(x)
    x = SepConv_BN(x, 728, "block4_sepconv1")
    x = SepConv_BN(x, 728, "block4_sepconv2", activation=False)
    x = layers.MaxPooling2D(3, strides=2, padding="same", name="block4_pool")(x)
    x = layers.add([x, residual], name="block4_add")

    for i in range(8):
        prefix = f"block{5+i}"
        residual = x
        x = layers.Activation("relu", name=f"{prefix}_act1")(x)
        x = SepConv_BN(x, 728, f"{prefix}_sepconv1")
        x = SepConv_BN(x, 728, f"{prefix}_sepconv2")
        x = SepConv_BN(x, 728, f"{prefix}_sepconv3", activation=False)
        x = layers.add([x, residual], name=f"{prefix}_add")

    residual = layers.Conv2D(1024, 1, strides=2, padding="same", use_bias=False, name="block13_res_conv")(x)
    residual = layers.BatchNormalization(axis=3, momentum=BN_MOMENTUM, epsilon=BN_EPS, scale=False, name="block13_res_bn")(residual)
    x = layers.Activation("relu", name="block13_act")(x)
    x = SepConv_BN(x, 728,  "block13_sepconv1")
    x = SepConv_BN(x, 1024, "block13_sepconv2", activation=False)
    x = layers.MaxPooling2D(3, strides=2, padding="same", name="block13_pool")(x)
    x = layers.add([x, residual], name="block13_add")

    x = SepConv_BN(x, 1536, "block14_sepconv1")
    x = SepConv_BN(x, 2048, "block14_sepconv2")

    x = layers.GlobalAveragePooling2D(name="avg_pool")(x)
    x = layers.Dropout(dropout, name="dropout")(x)
    out = layers.Dense(1, activation="sigmoid", name="pred")(x)
    return keras.Model(img_input, out, name="ManualXceptionBinary")

model = xception_build()
model.compile(
    optimizer=keras.optimizers.Adam(3e-4),
    loss="binary_crossentropy",
    metrics=[keras.metrics.BinaryAccuracy(name="accuracy"),
             keras.metrics.Precision(name="precision"),
             keras.metrics.Recall(name="recall")]
)
model.summary()

callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_loss", 
        patience=4,
        restore_best_weights=True,
        min_delta=1e-4,
        verbose=1
    )
]

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)


merged = pd.DataFrame(history.history)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(merged['accuracy'], label='train')
plt.plot(merged['val_accuracy'], label='val')
plt.title('Accuracy'); 
plt.xlabel('Epoch'); 
plt.ylabel('Accuracy'); 
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(merged['loss'], label='train')
plt.plot(merged['val_loss'], label='val')
plt.title('Loss'); 
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
plt.tight_layout(); 
plt.show()

class_names = ['background', 'logo']
y_true, y_pred = [], []
for images, labels in test_ds:
    probs = model.predict(images).ravel()
    y_true.extend(labels.numpy().astype(int))
    y_pred.extend((probs >= 0.5).astype(int))

acc = accuracy_score(y_true, y_pred)
prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
print(f"\nTEST metrics @0.5:  Accuracy={acc:.4f}  Precision={prec:.4f}  Recall={rec:.4f}  F1={f1:.4f}\n")

print("Classification Report (TEST):")
print(classification_report(y_true, y_pred, target_names=class_names, zero_division=0))

cm = confusion_matrix(y_true, y_pred, labels=[0,1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix (TEST)")
plt.show()


n_samples = 5
sample_idxs = random.sample(range(len(X_test)), n_samples)
plt.figure(figsize=(15, 4))

for i, idx in enumerate(sample_idxs, 1):
    path = X_test[idx]
    label_true = y_test[idx]
   
    img = cv2.imread(path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, (IMG_SIZE, IMG_SIZE))
    inp = (img_resized.astype("float32") / 255.0 - 0.5) * 2.0
    prob = float(model.predict(inp[None, ...], verbose=0)[0][0])
    label_pred = int(prob >= 0.5)

    plt.subplot(1, n_samples, i)
    plt.imshow(img_rgb)
    title = f"True: {class_names[label_true]}\nPred: {class_names[label_pred]} ({prob:.2f})"
    plt.title(title, fontsize=10, color=("blue" if label_true == label_pred else "red"))
    plt.axis("off")

plt.suptitle("Випадкові передбачення моделі", fontsize=14)
plt.tight_layout()
plt.show()


video_path = "/kaggle/input/vw-logo-2/video_vw/video_logo_vw.mp4"
intervals = detect_logo_periods_in_video(video_path, prob_threshold=0.5, frame_step=5)


def preprocess_frame_bgr(frame_bgr):
    img = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_LANCZOS4)
    img = img.astype("float32") / 255.0
    img = (img - 0.5) * 2.0
    return img

def merge_intervals(intervals, min_gap=0.5):
    if not intervals: return []
    intervals = sorted(intervals)
    merged = [list(intervals[0])]
    for s,e in intervals[1:]:
        if s - merged[-1][1] <= min_gap:
            merged[-1][1] = max(merged[-1][1], e)
        else:
            merged.append([s,e])
    return [tuple(x) for x in merged]

def detect_logo_periods_in_video(video_path, prob_threshold=0.5, frame_step=5, min_presence_seconds=0.2):
    assert os.path.exists(video_path), f"Відео не знайдено: {video_path}"
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError("Не вдалося відкрити відеофайл")

    fps = cap.get(cv2.CAP_PROP_FPS)
    intervals, current_on = [], None
    idx = 0

    while True:
        ret = cap.grab()
        if not ret: break
        if idx % frame_step == 0:
            ret2, frame = cap.retrieve()
            if not ret2: break
            img = preprocess_frame_bgr(frame)
            prob = float(model.predict(img[None, ...], verbose=0)[0][0])
            t_sec = idx / fps if fps > 0 else 0.0

            if prob >= prob_threshold:
                if current_on is None: current_on = t_sec
            else:
                if current_on is not None:
                    intervals.append((current_on, t_sec))
                    current_on = None
        idx += 1

    if current_on is not None:
        last_t = (idx-1)/fps if fps>0 else 0.0
        intervals.append((current_on, last_t))
    cap.release()

    intervals = [(s,e) for s,e in intervals if (e-s) >= min_presence_seconds]
    intervals = merge_intervals(intervals, min_gap=0.5)

    print(f"Found {len(intervals)} intervals at p>={prob_threshold} (frame_step={frame_step}).")
    for i,(s,e) in enumerate(intervals,1):
        print(f"  {i:02d}) {s:.2f}s — {e:.2f}s  (dur={e-s:.2f}s)")
    return intervals
```
